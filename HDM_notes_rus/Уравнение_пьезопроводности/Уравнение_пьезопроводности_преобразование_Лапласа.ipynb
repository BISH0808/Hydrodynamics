## Производительность вертикальной скважины в круговом пласте. Решение уравнения пьезопроводности с помощью преобразования Лапласа

Рассматривается двухпластовая система, постоянное забойное давление на скважине, постоянное давление на внешней границе пласта 1, отсутствие перетока на внешней границе пласта 2.

Уравнение пьезопроводности:
$$\frac{1}{\varkappa}\frac{\partial p}{\partial t}=\nabla^2 p$$

Распишем оператор Лапласа в цилиндрической системе координат (т.к. приток к скважине радиальный):

$$\frac{1}{\varkappa}\frac{\partial p(r,t)}{\partial t}=\frac{1}{r}\frac{\partial p(r,t)}{\partial r}+\frac{\partial^2 p(r,t)}{\partial r^2}$$

### Пласт 1

Начальное условие (пласт 1):
$$p(r,0)=0,\,\,\,r\in (r_w,r_{e1}]$$

Граничные условия (пласт 1 - постоянное давление на внешней границе):
$$p(r_w,t)=p_w-p_{e1}$$
$$p(r_{e1},t)=0$$

(в данном случае $p_{e1}$ -- давление, поддерживаемое на границе пласта 1, и начальное давление во всём пласте; для удобства последующего решения задачи из граничных и начальных условий вычли $p_{e1}$)

Перейдём в пространство Лапласа:
$$\begin{cases}
\displaystyle{}\frac{s}{\varkappa_1}=\frac{1}{r}\frac{\partial p}{\partial r}+\frac{\partial^2 p}{\partial r^2}\\\\
\displaystyle{}p(r_w)=\frac{p_w-p_{e1}}{s}\\
\displaystyle{}p(r_{e1})=0
\end{cases}$$

Фундаментальное решение в пространстве Лапласа:
$$p=C_1I_0\left(\sqrt{\frac{s}{\varkappa_1}}r\right)+C_2K_0\left(\sqrt{\frac{s}{\varkappa_1}}r\right)$$

И условия на константы $C_1$ и $C_2$:
$$\begin{cases}
\displaystyle{}C_1I_0\left(\sqrt{\frac{s}{\varkappa_1}}r_{e1}\right)+C_2K_0\left(\sqrt{\frac{s}{\varkappa_1}}r_{e1}\right)=0\\\\
\displaystyle{}C_1I_0\left(\sqrt{\frac{s}{\varkappa_1}}r_w\right)+C_2K_0\left(\sqrt{\frac{s}{\varkappa_1}}r_w\right)=\frac{p_w-p_{e1}}{s}
\end{cases}$$

Получаем выражения для констант

$$C_1=\frac{\dfrac{p_w-p_{e1}}{s}\cdot K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)}{K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)-I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)}$$

<br>

$$C_2=-\frac{\dfrac{p_w-p_{e1}}{s}\cdot I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)}{K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)-I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)}$$

<br>

и решение в пространстве Лапласа

$$\displaystyle{} p_1=\frac{\dfrac{p_w-p_{e1}}{s}\left[K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r\right)-I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r\right)\right]}{K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)-I_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_{e1}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_1}}r_w\right)}$$

### Пласт 2

Начальное условие (пласт 2):
$$p(r,0)=0,\,\,\,r\in (r_w,r_{e2}]$$

Граничные условия (пласт 2 - отсутствие перетока на внешней границе):
$$p(r_w,t)=p_w-p_{e2}$$
$$\frac{\partial p}{\partial r}\bigg|_{r=r_e}=0$$

(в данном случае $p_{e2}$ -- начальное давление во всём пласте; для удобства последующего решения задачи из граничных и начальных условий вычли $p_{e2}$)

Перейдём в пространство Лапласа:
$$\begin{cases}
\displaystyle{}\frac{s}{\varkappa_2}=\frac{1}{r}\frac{\partial p}{\partial r}+\frac{\partial^2 p}{\partial r^2}\\\\
\displaystyle{}p(r_w)=\frac{p_w-p_{e2}}{s}\\\\
\displaystyle{}\frac{\partial p}{\partial r}\bigg|_{r=r_e}=0
\end{cases}$$

Фундаментальное решение в пространстве Лапласа:
$$p=C_1I_0\left(\sqrt{\frac{s}{\varkappa_2}}r\right)+C_2K_0\left(\sqrt{\frac{s}{\varkappa_2}}r\right)$$

И условия на константы $C_1$ и $C_2$:

$$\begin{cases}
\displaystyle{}C_1I_1\left(\sqrt{\frac{s}{\varkappa_2}}r_{e2}\right)+C_2K_1\left(\sqrt{\frac{s}{\varkappa_2}}r_{e2}\right)=0\\\\
\displaystyle{}C_1I_0\left(\sqrt{\frac{s}{\varkappa_2}}r_w\right)+C_2K_0\left(\sqrt{\frac{s}{\varkappa_2}}r_w\right)=\frac{p_w-p_{e2}}{s}
\end{cases}$$

Получаем выражения для констант

$$C_1=\frac{\dfrac{p_w-p_{e2}}{s}\cdot K_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)}{K_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)-I_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)}$$

<br>

$$C_2=-\frac{\dfrac{p_w-p_{e2}}{s}\cdot I_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)}{K_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)-I_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)}$$

<br>

и решение в пространстве Лапласа

$$\displaystyle{} p_2=\frac{\dfrac{p_w-p_{e2}}{s}\left[K_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r\right)-I_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r\right)\right]}{K_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)I_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)-I_1\left(\sqrt{\dfrac{s}{\varkappa_2}}r_{e2}\right)K_0\left(\sqrt{\dfrac{s}{\varkappa_2}}r_w\right)}$$

```python
import sympy as sp
from anaflow import get_lap_inv
import matplotlib.pyplot as plt
import numpy as np
```

```python
s, r = sp.symbols('s, r')
```

```python
# значения параметров

mu1 = 0.0014  # [Па*с] вязкость жидкости пласта 1
mu2 = 0.0014  # [Па*с] вязкость жидкости пласта 2
k1 = 1e-14  # [м^2] проницаемость породы пласта 1
k2 = 1e-14  # [м^2] проницаемость породы пласта 2
h1 = 10  # [м] мощность (толщина) пласта 1
h2 = 5  # [м] мощность (толщина) пласта 2
phi1 = 0.2  # пористость породы пласта 1
phi2 = 0.1  # пористость породы пласта 2

rw = 0.1  # [м] радиус скважины
pw = 10132500  # [Па] начальное давление на скважине
re1 = 50  # [м] радиус кругового контура питания 1
pe1 = 25331250  # [Па] начальное давление на контуре питания пласта 1
re2 = 50  # [м] радиус кругового контура питания 2
pe2 = 25331250  # [Па] начальное давление на контуре питания пласта 2

Ct1 = 4.9345e-10  # [Па^(-1)] общая сжимаемость пласта 1
Ct2 = 4.9345e-10  # [Па^(-1)] общая сжимаемость пласта 2

kappa1 = k1/(phi1*mu1*Ct1)  # коэффициент пьезопроводности 1
kappa2 = k2/(phi2*mu2*Ct2)  # коэффициент пьезопроводности 2
```

```python
# аргументы функций Бесселя (пласт 1)
arg1_r, arg1_re1, arg1_rw = sp.sqrt(s/kappa1)*r, sp.sqrt(s/kappa1)*re1, sp.sqrt(s/kappa1)*rw
# аргументы функций Бесселя (пласт 2)
arg2_r, arg2_re2, arg2_rw = sp.sqrt(s/kappa2)*r, sp.sqrt(s/kappa2)*re2, sp.sqrt(s/kappa2)*rw

# распределение давления (пласт 1) в пространстве Лапласа
p1 = (pw - pe1)/s * \
    (sp.besselk(0, arg1_re1) * sp.besseli(0, arg1_r) - sp.besseli(0, arg1_re1) * sp.besselk(0, arg1_r)) / \
    (sp.besselk(0, arg1_re1) * sp.besseli(0, arg1_rw) - sp.besseli(0, arg1_re1) * sp.besselk(0, arg1_rw))

# распределение давления (пласт 2) в пространстве Лапласа
p2 = (pw - pe2)/s * \
    (sp.besselk(1, arg2_re2) * sp.besseli(0, arg2_r) + sp.besseli(1, arg2_re2) * sp.besselk(0, arg2_r)) / \
    (sp.besselk(1, arg2_re2) * sp.besseli(0, arg2_rw) + sp.besseli(1, arg2_re2) * sp.besselk(0, arg2_rw))
```

```python
# распределение давления (пласт 1) в пространстве Лапласа
p1
```

```python
# распределение давления (пласт 2) в пространстве Лапласа
p2
```

```python
# выражения для дебитов (из пласта 1 и из пласта 2) в пространстве Лапласа
q1 = ((2*sp.pi*k1*h1*rw)/mu1)*sp.diff(p1, r).subs(r, rw)
q2 = ((2*sp.pi*k2*h2*rw)/mu2)*sp.diff(p2, r).subs(r, rw)
```

```python
# преобразование символьных выражений в функции для расчётов с числовыми значениями
p1 = sp.lambdify([s, r], p1)
p2 = sp.lambdify([s, r], p2)
q1 = sp.lambdify(s, q1)
q2 = sp.lambdify(s, q2)
```

```python
time = np.linspace(36, 360000, 50)  # время в секундах

fig1, ax1 = plt.subplots(figsize=[10,6])

# цикл по разным радиусам
for _, ri in enumerate(np.linspace(rw, re1, 2)):
    # строим графики зависимостей давлений (найдены с помощью обратного преобразования Лапласа) от времени
    # в одном часе 3600 секунд и в одной атмосфере 101325 Па
    # +pe1 и +pe2, так как в самом начале решения вычитали их из граничных и начальных условий
    ax1.plot(time/3600, (get_lap_inv(p1, r=ri)(time) + pe1)/101325, ':', label=f'$p_1\,(r_1={round(ri,1)})$')
    ax1.plot(time/3600, (get_lap_inv(p2, r=ri)(time) + pe2)/101325, label=f'$p_2(r_2={round(ri,1)})$')

ax1.set_title('Зависимость давления от времени')
ax1.set_xlabel('t, ч')
ax1.set_ylabel('p, атм')
ax1.legend()
ax1.minorticks_on()
ax1.grid(which='major', linewidth=0.3)
ax1.grid(which='minor', linewidth=0.1)
```

```python
radius = np.linspace(rw, re1, 200)

fig2, ax2 = plt.subplots(figsize=[10,6])

# цикл по разным временам (неустановившегося режима)
for i, ti in enumerate(np.linspace(36, 3600, 5)):
    # строим распределения давлений (найдены с помощью обратного преобразования Лапласа) по радиусу
    # +pe1 и +pe2, так как в самом начале решения вычитали их из граничных и начальных условий
    ax2.plot(radius, (np.squeeze([get_lap_inv(p1, r=ri)(ti) for ri in radius]) + pe1)/101325, ':', label = f'$t={round(ti/3600, 1)} ч,пласт\,1$')
    ax2.plot(radius, (np.squeeze([get_lap_inv(p2, r=ri)(ti) for ri in radius]) + pe2)/101325, label = f'$t={round(ti/3600, 1)} ч, пласт\,2$')

ax2.set_title('Зависимость давления от радиуса (неустановившийся режим)')
ax2.set_xlabel('r, м')
ax2.set_ylabel('p, атм')
ax2.legend()
ax2.minorticks_on()
ax2.grid(which='major', linewidth=0.3)
ax2.grid(which='minor', linewidth=0.1)

#plt.tight_layout()
#plt.savefig('transient.png', format='png', dpi=300, facecolor='white')
```

```python
radius = np.linspace(rw, re1, 50)

fig2, ax2 = plt.subplots(figsize=[10,6])

# цикл по разным временам (установившегося режима пласта 1)
for i, ti in enumerate(np.linspace(36000, 360000, 5)):
    # строим распределения давлений (найдены с помощью обратного преобразования Лапласа) по радиусу
    # +pe1 и +pe2, так как в самом начале решения вычитали их из граничных и начальных условий
    ax2.plot(radius, (np.squeeze([get_lap_inv(p1, r=ri)(ti) for ri in radius]) + pe1)/101325, ':', label = f'$t={round(ti/3600, 1)} ч,пласт\,1$')
    ax2.plot(radius, (np.squeeze([get_lap_inv(p2, r=ri)(ti) for ri in radius]) + pe2)/101325, label = f'$t={round(ti/3600, 1)} ч, пласт\,2$')

ax2.set_title('Зависимость давления от радиуса')
ax2.set_xlabel('r, м')
ax2.set_ylabel('p, атм')
ax2.legend()
ax2.minorticks_on()
ax2.grid(which='major', linewidth=0.3)
ax2.grid(which='minor', linewidth=0.1)
```

```python
time = np.logspace(2, 7, 40)

# в одних сутках 86400 секунд
Q1 = 86400 * get_lap_inv(q1)(time)
Q2 = 86400 * get_lap_inv(q2)(time)

fig3, ax3 = plt.subplots(figsize=(10,6))
ax3.plot(time/3600, Q1, ':', label='$Q_1$')
ax3.plot(time/3600, Q2, '--', label='$Q_2$')
ax3.plot(time/3600, Q1 + Q2, label='$Q=Q_1+Q_2$')

ax3.set_xscale('log')
ax3.set_title('Зависимость дебита от времени')
ax3.set_xlabel('t, ч')
ax3.set_ylabel('$Q, м^3/сут$')
ax3.legend()
ax3.minorticks_on()
ax3.grid(which='major', linewidth=0.3)
ax3.grid(which='minor', linewidth=0.1)
```

```python
from matplotlib.animation import FuncAnimation
from IPython import display

radius = np.linspace(rw, re1, 200)

fig = plt.figure()
ax = plt.axes()
line1, = ax.plot([], label='Пласт 1', linewidth=0.7, color='blue')
line2, = ax.plot([], label='Пласт 2', linewidth=0.7, color='red')
ax.set_xlim(0, max(re1,re2))
ax.set_ylim(pw/101325, pe1/101325)
ax.set_xlabel('Радиус r, м')
ax.set_ylabel('Давление p, атм')
ax.grid(which='major', linewidth=0.3)
ax.grid(which='minor', linewidth=0.1)
ax.minorticks_on()

def animate(frame_num):
    line1.set_data((radius, (np.squeeze([get_lap_inv(p1, r=ri)(frame_num) for ri in radius]) + pe1)/101325))
    line2.set_data((radius, (np.squeeze([get_lap_inv(p2, r=ri)(frame_num) for ri in radius]) + pe2)/101325))
    ax.set_title(f'Распределение давления, время {round(frame_num/3600, 1)} ч')
    return line1, line2

anim = FuncAnimation(fig, animate, frames=np.concatenate([np.linspace(1,3600,100), np.linspace(3600,72000,100)]),
                     interval=100, repeat=False)
plt.legend()
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
